## Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('../dataset/stocks.csv')


df.head()
```

## Data Cleaning

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

## Exploratory Data Analysis

```python
df['Ticker'].unique()
```

```python
df.groupby('Ticker')[['Open','Close']].mean()
```

```python
df.groupby('Ticker')['Volume'].mean().sort_values(ascending=False)
```

## Feature Engineering

```python
df['Daily_Return'] = ((df['Close'] - df['Open']) / df['Open']) * 100
df[['Ticker','Daily_Return']].head()
```

## Data Visualization

```python
plt.figure(figsize=(12,6))

for ticker in df['Ticker'].unique():
    subset = df[df['Ticker'] == ticker]
    plt.plot(subset['Date'], subset['Close'], label=ticker)

plt.title('Closing Price Trend by Company')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.show()
```

```python
avg_close = df.groupby('Ticker')['Close'].mean()

plt.figure(figsize=(8,5))
avg_close.plot(kind='bar')

plt.title('Average Closing Price by Company')
plt.xlabel('Company')
plt.ylabel('Average Close Price')
plt.show()
```

```python
plt.figure(figsize=(8,5))

df['Volume'].hist(bins=20)

plt.title('Trading Volume Distribution')
plt.xlabel('Volume')
plt.ylabel('Frequency')
plt.show()
```
